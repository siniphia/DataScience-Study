# 3. DataFrame 생성하기

## (1) Dictionary

```python
dictFriends = [
    {'name' : 'Lee', 'age' : 25, 'university' : 'Chung-Ang', 'major' : 'CS'},
    {'name' : 'David', 'age' : 28, 'university' : 'Stanford', 'major' : 'EE'},
    {'name' : 'Lee', 'age' : 25, 'university' : 'Chung-Ang', 'major' : 'CS' }
]
```

```python
myFriends = pd.DataFrame(dictFriends)
myFriends
```

```python
# dict : column을 수동정렬 해야 함
myFriends = myFriends[['name', 'age', 'university', 'major']]
myFriends
```

## (2) List

```python
listFriends = [
    ['Lee', 25, 'Chung-Ang', 'CS'],
    ['David', 28, 'Stanford', 'EE'],
    ['Kim', 22, 'Seoul', 'BE']
]

columnName = ['name', 'age', 'university', 'major']
```

```python
myFriends = pd.DataFrame.from_records(listFriends, columns = columnName)
myFriends
```

# 5. 행, 열 선택 및 필터

```python
mushroom.head()
```

## (1) Row Selection
+ **참조사이트** : http://yeyej.blogspot.com/2016/02/pandas-dataframe-iloc-loc-ix.html
+ **loc, iloc, ix**

```python
# loc : label에 의한 Selection
mushroom.loc[[0, 3, 9]]
```

```python
mushroom.loc[:4] # 값이 4인 애들까지 출력
```

```python
# iloc : index에 의한 Selection
mushroom.iloc[:4] # index가 3인 애들까지 출력
```

## (2) Row Selection by Condition

```python
# 단일조건
temp = mushroom[mushroom.odor=='p']
temp.head()
```

```python
# 복합조건
temp = mushroom[(mushroom.bruises=='t') & (mushroom.odor=='p')]
temp.head()
```

## (3) Column Selection by Filter
+ **items, like, regex**

```python
temp = mushroom.filter(items = ['odor', 'population'])
temp.head()
```

```python
temp = mushroom.filter(like = 'cap', axis = 1)
temp.head()
```

```python
temp = mushroom.filter(regex='color$', axis = 1)
temp.head()
```

# 6. 행, 열 삭제

## (1) Row Deletion

```python
# drop : 해당 label의 row를 제거함
temp = mushroom.drop([0, 3, 5])
temp.head()
```

```python
# inplace : 리턴값이 바로 적용됨 -> 따로 대입할 필요 없음
temp.drop([2, 4, 6], inplace = True)
temp.head()
```

```python
# index : 해당 label이 아닌 index의 row를 제거함
temp.drop(temp.index[:4], inplace = True)
temp.head()
```

## (2) Column Deletion

```python
# axis = 1 : Column을 의미!!
temp = temp.drop('cap-surface', axis=1)
temp.head()
```

# 7. 행, 열 생성 및 수정

## (1) Column Creating

```python
# df['name'] : 해당 name의 column을 dataframe에 추가
mushroom['price'] = 1
mushroom.head()
```

```python
# np.where을 활용해 새 Column에 Value Assign
# Q) where문 안에 '복합조건'을 사용할 수 없는가?
mushroom['rarity'] = np.where(mushroom['habitat'] == 'u', 'high', 'low')
mushroom.head()
```

## (2) Column Modifying

```python
# 먹을 수 있으면 높게, 그렇지 않으면 낮게 가치측정
def price_calculating(row) :
    if row == 'e':
        return random.randint(30, 80)
    else:
        return random.randint(0, 10)
```

```python
# class column에 함수 apply해서 price column을 생성하고 값 삽입
mushroom['price'] = mushroom['class'].apply(price_calculating)
```

```python
mushroom.head()
```

## (3) Row Creating

```python
students = [
    ['James', '25', '100', '87'],
    ['Kafka', '32', '80', '79'],
    ['Viera', '27', '90', '65'],
    ['Phobe', '18', '60', '78']
]
column_name = ['Name', 'Age', 'Midterm', 'Final']
```

```python
df = pd.DataFrame.from_records(students, columns = column_name)
```

```python
df.head()
```

```python
new_student = [['Keane', '28', '80', '90']]
df2 = pd.DataFrame.from_records(new_student, columns = column_name)
df2
```

----------

```python
df = df.append(df2, ignore_index = True)
df
```

# 8. 중복 Row 삭제

## (1) 겹치는 모든 Row 삭제

```python
# 중복된 Row 추가하기
df = df.append(df2, ignore_index = True)
df
```

```python
# Duplicated('column') : 해당 컬럼값이 중복되는 row가 있으면 True 반환
df.duplicated('Name')
```

```python
# Duplicates() : 중복된 row 제거!!
df = df.drop_duplicates()
```

## (2) 특정 Column이 같은 Row 삭제

```python
new_student = [['Phobe', '18', '60', '78']]
df2 = pd.DataFrame.from_records(new_student, columns = column_name)
df2
```

```python
df = df.append(df2, ignore_index = True)
df
```

```python
# 이름이 같은걸 제거, keep = first -> 처음걸 남김, last -> 나중걸 남김
df = df.drop_duplicates(['Name'], keep = 'first')
df
```
