# 3. DataFrame 생성하기

## (1) Dictionary

```python
dictFriends = [
    {'name' : 'Lee', 'age' : 25, 'university' : 'Chung-Ang', 'major' : 'CS'},
    {'name' : 'David', 'age' : 28, 'university' : 'Stanford', 'major' : 'EE'},
    {'name' : 'Lee', 'age' : 25, 'university' : 'Chung-Ang', 'major' : 'CS' }
]
```

```python
myFriends = pd.DataFrame(dictFriends)
myFriends
```

```python
# dict : column을 수동정렬 해야 함
myFriends = myFriends[['name', 'age', 'university', 'major']]
myFriends
```

## (2) List

```python
listFriends = [
    ['Lee', 25, 'Chung-Ang', 'CS'],
    ['David', 28, 'Stanford', 'EE'],
    ['Kim', 22, 'Seoul', 'BE']
]

columnName = ['name', 'age', 'university', 'major']
```

```python
myFriends = pd.DataFrame.from_records(listFriends, columns = columnName)
myFriends
```

----------
# 4. DataFrame 파일로 저장하기

```python
myFriends.to_csv('friends.csv', index = True, header = True)
```

# 5. 행, 열 선택 및 필터

```python
mushroom.head()
```

## (1) Row Selection
+ **참조사이트** : http://yeyej.blogspot.com/2016/02/pandas-dataframe-iloc-loc-ix.html
+ **loc, iloc, ix**

```python
# loc : label에 의한 Selection
mushroom.loc[[0, 3, 9]]
```

```python
mushroom.loc[:4] # 값이 4인 애들까지 출력
```

```python
# iloc : index에 의한 Selection
mushroom.iloc[:4] # index가 3인 애들까지 출력
```

## (2) Row Selection by Condition

```python
# 단일조건
temp = mushroom[mushroom.odor=='p']
temp.head()
```

```python
# 복합조건
temp = mushroom[(mushroom.bruises=='t') & (mushroom.odor=='p')]
temp.head()
```

## (3) Column Selection by Filter
+ **items, like, regex**

```python
temp = mushroom.filter(items = ['odor', 'population'])
temp.head()
```

```python
temp = mushroom.filter(like = 'cap', axis = 1)
temp.head()
```

```python
temp = mushroom.filter(regex='color$', axis = 1)
temp.head()
```
